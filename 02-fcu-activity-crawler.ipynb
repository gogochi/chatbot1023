## 搜尋標籤

find and find_all  
https://www.crummy.com/software/BeautifulSoup/bs4/doc/#searching-the-tree  
find(tag, attributes, recursive, string, keywords)  
find_all(tag, attributes, recursive, string, limit, keywords)  

**find() 只回傳第一個找到的區塊，而 find_all() 會回傳一個 list, 包含所有符合條件的區塊。**

95% 爬蟲，只會用到tag 跟 attributes  

The string argument is new in Beautiful Soup 4.4.0. In earlier versions it was called text  

```python
#抓table
table = soup.find('table',{'class':'tb'})
table.text
```

```python
# 抓row
for row in table.find_all('tr'):
    print(row.text)
```

```python
t = table.find('tr')
for row in t.find_all('th'):
    print(row.text)
```

```python
#抓表格欄位標題
for title in table.find('tr').find_all('th'):
    print(title.text)
```

```python
# 抓欄位值
for row in table.find_all('tr')[1:]:
    fields = row.find_all('td')
    print(fields[0].text,fields[1].text,fields[2].text,fields[3].text)
```

```python
# 加入連結
for row in table.find_all('tr')[1:]:
    fields = row.find_all('td')
    link = 'http://news.fcu.edu.tw/wSite/'+fields[1].find('a', href=True).get('href')
    
    post_date = fields[0].text.strip()
    title = fields[1].text.strip()
    start_date = fields[2].text.strip()
    end_date = fields[3].text.strip()
    print(post_date,title,start_date,end_date)
    print(link)
```

## 校園新聞
http://news.fcu.edu.tw/wSite/lp?ctNode=15360&mp=9000

```python
# your code
# 校園新聞
import requests
from bs4 import Tag, BeautifulSoup as bs
r = requests.get("http://news.fcu.edu.tw/wSite/lp?ctNode=15360&mp=9000&pagesize=50")
soup = bs(r.text,"lxml")
table = soup.find('table',{'class':'tb'})

#抓表格欄位標題
for title in table.find('tr').find_all('th'):
    print(title.text)
    
# 加入連結
for row in table.find_all('tr')[1:]:
    fields = row.find_all('td')
    link = 'http://news.fcu.edu.tw/wSite/'+fields[1].find('a', href=True).get('href')
    
    post_date = fields[0].text.strip()
    title = fields[1].text.strip()
    print(post_date,title)
    print(link)
```
